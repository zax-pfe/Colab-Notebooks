comme il y a des maison trop cheres qui faussent la representation, on va degager les 1% des maisons les plus chere 

```
non_top_1_perc = df.sort_values('price', ascending=False).iloc[216:]
non_top_1_perc.shape
```

edgecolor signifie que nous allons pas tracer le bord de nos points, alpha nous allons les rendre transparents, et palette signifie que nous allons donner des couleurs allant du red au yelow au green

```
plt.figure(figsize=(12,8))
sns.scatterplot(x='long', y='lat', data=non_top_1_perc, hue='price', edgecolor=None, alpha=0.2, palette='RdYlGn')
```
