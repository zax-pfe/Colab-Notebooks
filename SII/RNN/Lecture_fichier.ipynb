```python
path = 'Fonction_1_SBT3.txt'
new_path = 'new_sbt.txt'
```

```python

sbt = open(path,'r').read()
print(len(sbt))
liste = []
for i in sbt:
  liste.append(i)
```

```python
print(liste)
longueur = len(liste)
```

```python
counter=0
for i in liste:
  if (i==')') or (i=='('):
    counter = counter+1

print(counter)
slots = 2*counter
slots
```

```python
i=0
while ( i < longueur+slots):
  if (liste[i] == '(') or (liste[i] == ')'):
    liste.insert(i," ")
    liste.insert(i+2," ")
    i=i+2
  i=i+1

```

```python
print(liste)
len_new_liste = len(liste)
```

```python
new_sbt = open(new_path,'a')

for i in range(len_new_liste):
  new_sbt.write(liste[i])
```
