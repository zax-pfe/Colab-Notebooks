```python
bdd,test = return_bdd_sbt()
```

```python
def generate_template(code):
    txt_trous = return_txt_trous_unique(code)
    encoded = encode_unique(txt_trous)
    generated_template = return_generated_template(encoded)
    return generated_template
```

```python
def generate_var(code):
    variables = return_variables_unique(code)
    encoded_var = encode_unique_var(variables)
    generated_var = return_generated_var(encoded_var)
    return generated_var
```

```python
def return_prediction(code):
    template = generate_template(code)
    #print(' '.join(template))
    predicted_var = generate_var(code)
    #print(predicted_var)

    j=0
    for i,word in enumerate (template):
        if word == '<oov>':
            template[i] = predicted_var[j]
            j=j+1    
    return ' '.join(template)
```

```python
for i in range(3):
    a = randint(0,5000)
    print('\n\nPredicted BDD : \n',return_prediction(test[a]))
    print('\nExpected BDD : \n', bdd[a],'\n\n')
```
