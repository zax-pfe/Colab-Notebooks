# Conversion en one hot encoding, dimension ( nombre fichiers : 1000, nombre max de caractere : 1572, nombre de caracteres differents : 49 bdd, 54 sbt )

# Initialisation des matrices de One Hot encoding

```python
encoded_bdd = np.zeros((number_of_folder,1572,54))
encoded_sbt = np.zeros((number_of_folder,1572,54))
```

# Conversion du sbt et bdd en one hot encoding

```python
for i,bdd_code in tqdm( enumerate(bdd)):
    for j,char in enumerate (bdd_code):
        encoded_bdd[i,j,char_to_ind_bdd[char]] = 1.
```

```python
for i,sbt_code in tqdm( enumerate(sbt)):
    for j,char in enumerate (sbt_code):
        encoded_sbt[i,j,char_to_ind_sbt[char]] = 1.
```

```python
encoded_bdd.shape
```

```python
encoded_sbt.shape
```

# Decodage du one hot encoding

```python
decoded_sequence_bdd = []
encoded_sequence_bdd = encoded_bdd[0]
for i,chars in enumerate(encoded_sequence_bdd):
    for j,char in enumerate(chars):
        if (char==1):
            decoded_sequence_bdd.append(ind_to_char_bdd[j])
decoded_sequence_bdd = ''.join(decoded_sequence_bdd)
```

```python
decoded_sequence_sbt = []
encoded_sequence_sbt = encoded_sbt[0]
for i,chars in enumerate(encoded_sequence_sbt):
    for j,char in enumerate(chars):
        if (char==1):
            decoded_sequence_sbt.append(ind_to_char_sbt[j])
decoded_sequence_sbt = ''.join(decoded_sequence_sbt)
```

```python
print(decoded_sequence_bdd)
print(decoded_sequence_sbt)
```

```python
dataset = tf.data.Dataset.from_tensor_slices((encoded_sbt,encoded_bdd))
```

# Creation du Modele

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GRU, Dropout, TimeDistributed, LSTM, Bidirectional, Embedding, RepeatVector,Input
from tensorflow.keras.losses import sparse_categorical_crossentropy
```

```python
model = Sequential()
#model.add(Input(shape=(1572, 54)))
model.add(Embedding(len(vocab_sbt),256))
#model.add(Embedding(len(vocab_sbt),256,input_length=1572))
#model.add(Bidirectional(LSTM(256, return_sequences=True)))
model.add(Bidirectional(LSTM(256, return_sequences=True)))
model.add(Dense(1024, activation='relu'))
print(model.input_shape)
model.add(Dropout(0.5))
model.add(Dense(54, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```
